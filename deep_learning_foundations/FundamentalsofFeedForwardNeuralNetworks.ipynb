# 2 - Introduction to the Artifical Neural Network

The concept of neural networks is inspired by the study of how our brain learns new information. Researchers in the field of cognitive science have learned that through a mechanism called neural plasticity, the brain can learn and adapt to new information. In the 1940s and 1950s, researchers in artificial intelligence started developing mathematical models designed to mimic the behavior of neurons and the way they process information during neural plasticity. These models are known as *Artificial Neural Networks (ANNs)*. We will be learning about *Feed-Forward Networks (FFN)*, the most basic ANN architecture.

## A. Basic Architecture of an FFN

The architecture of an FFN typically consists of three main layers:

* **Input Layer:** The input layer receives the input data and passes it on to the next layer. The number of neurons in the input layer corresponds to the number of features in the input data.

* **Hidden Layers:** The hidden layers process the input data and perform computations on it. The number of hidden layers and the number of neurons in each layer can vary depending on the complexity of the problem. In general, the more hidden layers and neurons, the more powerful the network will be in capturing complex patterns in the data.

* **Output Layer:** The output layer produces the final output of the network. The number of neurons in the output layer corresponds to the number of output classes or output values.

A feed-forward network is called "deep" when there is more than one hidden layer in the network.


## B. Weights and Biases

Every neuron in one layer is connected to the neurons in the next layer. They are connected through a weight value. *Weights* can be represented as a matrix that will be multiplied by the neurons at the beginning of the link, which can also be represented as a matrix, to get the raw outputs of that link. This is modeled using matrix multiplication.

*Biases*, on the other hand, are additional parameters in neurons that are used to shift the activation function of the neuron. They allow the network to represent patterns even when all inputs are zero or close to zero. Biases provide flexibility and enable the model to better fit the training data.

The process of multiplying inputs by weights, and adding the bias to it is called a linear transformation. Weights and biases are essentially what we need to find the optimal value for that will give us the best predictions. We will discuss more about how they are used to predict, and how we update the values when we talk about the training process.

After every layer linear transformation, a non-linear transformation called an *activation function* is applied. This helps to introduce non-linearity into the network. # 4 - Training an FFN

When training an FFN, there are two processes ongoing sequentially back and forth:

1. **Forward Propagation (Forward Pass)** - Forward propagation is where the model goes left-to-right, where it takes the inputs and goes through all the layers to recieve an output. Essentially, the forward pass makes a prediction with the current weight and bias values. We use this prediction to compare with our real values and calculate a loss value using a *loss function*.

2. **Backpropagation (Backward Pass)** - The backward pass is how we update the parameters of a neural network to minimize our loss value, going right-to-left. The goal of the backward pass is to essentially update our weights and biases layer by layer using an algorithm used to minimize our loss, helping us converge to the optimal solution.

## A. Forward Propagation




### I. Forward Pass Algorithm  
During the Forward Propagation phase, the input data is multiplied by the weights of the connections between the neurons in each layer. This is done using matrix multiplication, which gives us a weighted sum. Then, the bias value of the neuron is added to this sum. The resulting value is then passed through an activation function.  
  
Below is the full algorithm for forward propagation:  
  
$$\begin{align*}  
\text{Input:} & \quad \mathbf{x} \quad \text{(input data matrix)} \\  
\text{Output:} & \quad \mathbf{a}^{[l]} \quad \text{(activations of the output layer)} \\  
\text{1.} & \quad \text{Initialize } \mathbf{a}^{[0]} \text{ as } \mathbf{x} \\  
\text{2.} & \quad \text{For } l = 1 \text{ to } l: \\  
\text{3.} & \quad \quad \mathbf{z}^{[l]} = \mathbf{a}^{[l-1]} \cdot \mathbf{w}^{[l]}  + \mathbf{b}^{[l]} \\  
\text{4.} & \quad \quad \mathbf{a}^{[l]} = a^{[l]}(\mathbf{z}^{[l]}) \\  
\text{5.} & \quad \text{Return } \hat{y} := \mathbf{a}^{[l]} \text{ (Output layer activation)}  
\end{align*}$$  
  
Where:  
* $l$ represents the number of layers in a network.  
* $\mathbf{x}$ represents the input data matrix. Shape should be $(\text{n_samples}, \text{n_input_features})$.  
* $\mathbf{a}^{[l]}$ and $\hat{y}$ denote the activations of the output layer. Shape should be $(\text{n_samples}, \text{n_outputs})$.  
* $\mathbf{a}^{[0]}$ is initialized as $\mathbf{x}$.  
* $\mathbf{z}^{[l]}$ is the linear activation calculated as $\mathbf{a}^{[l-1]} \cdot \mathbf{w}^{[l]}  + \mathbf{b}^{[l]}$ in step 3. We denote the value of $\mathbf{z}^{[l]}$ as a function $z^{[l]}$.  
* $\mathbf{w}^{[l]}$ represents the weight matrix of layer $l$.  
* $\mathbf{b}^{[l]}$ is the bias vector of layer $l$.  
* $a^{[l]}(\cdot)$ denotes the activation function applied to $\mathbf{z}^{[l]}$ in step 4.  


### II. Loss Calculation

After we have our predictions, we can calculate the loss value using a loss function $C(\hat{y}_i, y_i)$. The goal is to minimize the value of this loss function every iteration. Below are common cost functions used:

1. **Mean Squared Error (MSE)** - Measures the average of the squares of the errors or deviations. It's commonly used for regression problems.
$$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

    Where:
    * $n$: Number of samples
    * $y_i$: True value of the i-th sample
    * $\hat{y}_i$: Predicted value of the i-th sample
2. **Binary Cross-Entropy Loss (or Log Loss)** - Used for binary classification problems. It penalizes predictions that are far from the actual class.
$$ BCE = -\frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
 $$

    Where:
    * $n$: Number of samples
    * $y_i$: True label of the i-th sample (0 or 1)
    * $\hat{y}_i$: Predicted probability of the i-th sample being in class 1
3. **Categorical Cross-Entropy Loss** - Utilized for multi-class classification problems. Each instance can belong to one class out of multiple classes.
$$ CCE =  -\frac{1}{n} \sum_{i=1}^{n} \sum_{j=1}^{m} y_{ij} \log(\hat{y}_{ij}) $$

    Where:
    * $n$: Number of samples
    * $m$: Number of classes
    * $y_i$: True one-hot encoded label of the i-th sample
    * $\hat{y}_i$: Predicted probability distribution over classes for the i-th sample
4. **Sparse Categorical Cross-Entropy Loss** - Similar to Categorical Cross-Entropy Loss, but expects integer target labels instead of one-hot encoded targets.
$$ SCCE = - \frac{1}{n} \sum_{i=1}^{n} \log(\hat{y}_{i}[y_i]) $$

    Where:
    * $n$: Number of samples
    * $y_i$: True label of the i-th sample (integer)
    * $\hat{y}_i$: Predicted probability distribution over classes for the i-th sample







## B. Backpropagation

Backpropagation is the method used to calculate the gradient of the loss function with respect to the parameters (weights and biases) of the neural network. It involves two main steps: the computation of the gradients with respect to the parameters in each layer, and the update of these parameters using an optimization algorithm (e.g., gradient descent).






### I. Compute Gradients  
The first step in backpropagation is to calculate gradients for each parameter. The *gradient* represents the rate of change of the loss function with respect to the weights and biases. It tells us how much the loss function will change if we make a small change to the weights. Mathematically, the gradient of a function with respect to a variable is a vector that points in the direction of the steepest increase of the function at a particular point.  
  
We calculate gradients by applying the chain rule from calculus. This allows us to decompose the gradient of the composite function into the product of the gradients of its individual components. In the case of a neural network, this means we compute the gradient of the loss function with respect to each parameter θ, represented as the partial derivative $\frac{∂C}{∂θ^{[l]}}.$ Below shows how each gradient is calculated:  
  
$$\frac{∂C}{∂θ^{[l]}} = \frac{∂C}{\partial z^{[l]}} \cdot \frac{∂ z^{[l]}}{\partial θ^{[l]}}$$  
  
Where:  
* $\frac{∂C}{\partial z^{[l]}}$ is the error for the unactivated output for layer $l$. We propagate this error back to compute the previous layer's error, and so on.  
* $\frac{∂ z^{[l]}}{\partial θ^{[l]}}$ is the derivative of the function $z^{[l]}$ that computes the array $z^{[l]}$.  
  
When expanding the chain rule of this formula, the formula begins to get bigger as we take many partial derivatives to compute these values. For example, if we have an activation function $a^{[l]}$, in order to compute $\frac{∂C}{\partial z^{[l]}}$, we need to take the partial derivative of $a^{[l]}$ with respect to the array $z^{[l]}$:  
  
$$\frac{∂C}{\partial z^{[l]}} = \frac{∂C}{\partial a^{[l]}} \cdot \frac{∂a^{[l]}}{\partial z^{[l]}}$$  
  
Let's expand it a bit to view the values of the chain rule computation:  
  
$$\begin{align*}  
\text{1.} & \quad \text{Initialize gradients:} \\  
& \quad \quad \frac{∂C}{\partial z^{[l]}} = \frac{\partial C}{\partial a^{[l]}} \odot \frac{∂a^{[l]}}{\partial z^{[l]}} \\  
\text{2.} & \quad \text{Loop backward through layers:} \\  
& \quad \text{For } l = L, L-1, \ldots, 1: \\  
\text{3.} & \quad \quad \text{Compute gradients:} \\  
& \quad \quad \quad \frac{∂C}{\partial \mathbf{w}^{[l]}} =  (\mathbf{a}^{[l-1]})^T \cdot \frac{∂C}{\partial z^{[l]}} \\  
& \quad \quad \quad \frac{∂C}{\partial \mathbf{b}^{[l]}} =  \sum_{i=1}^{m} \frac{∂C}{\partial z^{[l]}} \\  
& \quad \quad \quad \frac{∂C}{\partial z^{[l-1]}} = (\frac{∂C}{\partial z^{[l]}} \cdot (\mathbf{w}^{[l]})^T) \odot \frac{∂a^{[l-1]}}{\partial z^{[l-1]}} \\  
\text{4.} & \quad \text{Return gradients:} \\  
& \quad \quad \frac{∂C}{\partial \mathbf{w}^{[l]}}, \frac{∂C}{\partial \mathbf{b}^{[l]}}: \quad ∀ l ∈ L\\  
\end{align*}$$  
  
Where:  
- $m$ is the number of samples.  
- $\mathbf{a}^{[0]} = \mathbf{x}$.  
- $\frac{∂a^{[l]}}{\partial z^{[l]}}$ denotes the derivative of the activation function applied to $z^{[l]}$.  
- $\odot$ denotes element-wise multiplication.  
- $C$ is the loss function with respect to the output layer activations $a^{[l]}$.  
- $\frac{∂C}{\partial z^{[l]}}$ is the derivative of the loss with respect to the linear activation of the output layer.  


### II. Gradient Descent

Now that we have computated the value of the gradients, we need a way to update our current parameters with some optimizer.One popular optimization algorithm is *Stochastic Gradient Descent (SGD)*. However, there are many popular algorithms that are also used, including *Adam*, *RMSProp*, *Adagrad*, *AdaDelta*, and *AdamW*. Adam and its varients has proven to be the most effective in many cases. However, we will use SGD to show how a basic optimizer works.

We introduce a *learning rate* parameter $\alpha$ that controls the speed of optimization. The common range for learning rates typically falls between 0.001 to 0.1, but this can vary depending on the problem and dataset. It is important to note that choosing a value for $\alpha$ too high  the optimization algorithm may overshoot the minimum of the loss function and fail to converge, and a $α$ too low will cause the training process to be very slow, or the model may get stuck in a local minima and struggle to escape it due to the small step size. Choosing an appropriate learning rate is important for optimal convergence.

Below is the algorithm for SGD:



$$
\textbf{W}^{[l]} := \textbf{W}^{[l]} - \alpha \cdot \frac{∂C}{\partial \mathbf{W}^{[l]}}
$$

$$
\textbf{b}^{[l]} := \textbf{b}^{[l]} - \alpha \cdot \frac{∂C}{\partial \mathbf{b}^{[l]}}
$$

Where:

* $\textbf{W}^{[l]}$: The weight matrix for layer $l$.
* $\textbf{b}^{[l]}$: The bias matrix for layer $l$.
* $\alpha$: The learning rate.
* $\frac{∂C}{\partial \mathbf{W}^{[l]}}$: The gradient weight matrix for layer $l$.
* $\frac{∂C}{\partial \mathbf{b}^{[l]}}$: The gradient bias matrix for layer $l$.

Gradient descent looks to minimize the error step by step incrementally by adjusting the weights and biases by a portion of the gradient for each parameter. Below is a visualization by [Hakob Avjyan](https://medium.com/@hakobavjyan/stochastic-gradient-descent-sgd-10ce70fea389) to view this:

## C. Mini-Batch Processing

In deep learning training, we iterate our data for each training step through mini-batches. We do this because:

1. **Efficiency**: Training deep neural networks on large datasets can be computationally intensive. Mini-batch processing allows us to process smaller subsets (mini-batches) of the data at a time, reducing memory requirements and speeding up computation compared to processing the entire dataset in one go.

2. **Regularization**: Mini-batch processing introduces noise into the optimization process. This noise can act as a form of regularization, helping to prevent overfitting and improve the generalization ability of the model.

3. **Gradient Descent**: Mini-batch processing is essential for implementing gradient descent optimization algorithms. These algorithms update the model parameters based on gradients computed from mini-batches of the training data.

4. **Parallelization**: Mini-batch processing enables parallelization of computation across multiple processors or devices, such as GPUs or distributed computing frameworks. Each mini-batch can be processed independently, allowing for more efficient use of computational resources.

5. **Convergence**: Mini-batch processing can lead to faster convergence during training compared to batch processing, especially for large datasets. By updating the model parameters more frequently, mini-batch processing can help the model converge to a solution more quickly.

When inputing data into the forward pass of an FFN, the shape should be (batch_size, n_features), therefore making the output shape (batch_size, n_outputs).

Lets look at an example with a random dataset with 100 samples and 5 columns:

```python
import numpy as np

dataset = np.random.randint(0, 100, size=(100, 5))
dataset.shape
```

When doing your mini-batch creation, you will use frameworks like PyTorch or TensorFlow to batch your data. However, since we have not introduced these frameworks, I will design a basic batch generator with just NumPy:

```python
class BatchGenerator:
    def __init__(self, data, batch_size):
        self.data = data
        self.batch_size = batch_size
        self.num_samples = len(data)
        self.num_batches = int(np.ceil(self.num_samples / self.batch_size))
        self.current_batch = 0

    def __len__(self):
        return self.num_batches

    def __iter__(self):
        return self

    def __next__(self):
        start_idx = self.current_batch * self.batch_size
        end_idx = min(start_idx + self.batch_size, self.num_samples)

        if start_idx >= self.num_samples:
            self.current_batch = 0
            raise StopIteration

        batch = self.data[start_idx:end_idx]

        self.current_batch += 1

        return batch
```

Let us view the shapes of each of the batches:

```python
data_loader = BatchGenerator(dataset, batch_size=20)
print(f'Number of Batches: {len(data_loader)}')
for i, batch in enumerate(data_loader):
    print(f'Batch {i+1}: {batch.shape}')
```

If we were to train this model, each forward pass would take in a shape of (batch_size, n_features), which is (20, 5) in this case.

## D. Epochs

As mentioned above, we iterate through mini-batches, until we do the forward pass, backward pass, and gradient descent of each mini-batch. Everytime we reach the end of one loop of this, we call it an *epoch*. We iterate over our all our batches for a certain amount of epochs until convergence. We must choose a value for epochs that will allow our model to converge with overfitting or underfitting. We will discuss overfitting and underfitting later in the course, and how to deal with these issues.

## E. Minotoring Optimality

As mentioned earlier, the goal is minimize the value of our loss function. However, this does not mean we monitor how efficient our model is based on just the loss value during training for each epoch.

We instead use a validation dataset to monitor loss. We do this because our validation set represents unseen data, meaning it will have a better representation of how our model reacts to data it has not been trained on. We want to make sure the validation loss for each epoch decreases for every epoch.

In addition to monitoring the loss value, we can also use metrics to better make sense of our model. For example, if we have a binary classification model, we can use accuracy as a metric to see how accurate our model is on average for each epoch. Monitoring validation metrics and losses can best describe the improvement epoch-to-epoch.



## F. Full Training Loop

Below I will show the entire basic training loop for training an FFN.

$$
\begin{align*}
& \text{For } \text{each } \text{epoch} \text{ in } \text{range(epochs)}: \\
& \quad\quad \text{For } \text{each } \text{batch} \text{ in } \text{data_loader}: \\
& \quad\quad\quad\quad \text{Forward pass}: \quad \textbf{output} = \textbf{forward_pass}(\textbf{batch}) \\
& \quad\quad\quad\quad \text{Compute loss}: \quad J = J(\textbf{output}, \textbf{target}) \\
& \quad\quad\quad\quad \text{Backward pass}: \quad \nabla_{\theta} J = \textbf{backward_pass}(J) \\
& \quad\quad\quad\quad \text{Gradient descent}: \quad \theta \leftarrow \theta - \alpha \nabla_{\theta} J \\
\end{align*}
$$

Every epoch, our model should come closer and closer to optimal convergence.

# 5 - Common Activation Functions

As mentioned before, activation functions serve the purpose of adding non-linearity to the hidden layers of the model. They are also used to transform outputs of the forward pass into proper format. For example, for a multiclass classification problem, we typically apply and activation function to convert raw predictions (logits) into probabilies.

Lets go over some of the common activation functions below.

## A. Rectified Linear Unit (ReLU)

*ReLU* is an activation function that outputs $x$ if it is greater than zero. Otherwise, if it is below zero, it will just output zero. This introduces sparse representations wich can reduce redundacy in the learned features. Below are the function for ReLU.

$$
\text{ReLU}(x) = \max(0, x)
$$

### I. Visualizing ReLU

```python
import numpy as np
import matplotlib.pyplot as plt


def relu(x):
    """
    ReLU Activation Function
    """
    return np.maximum(x, 0)


# Make linear array
x = np.linspace(-10, 10, 100)

# Init subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Plot Unactivated
ax[0].plot(x, x)
ax[0].set_title('Unactivated')

ax[0].set_ylim(x.min()-1, x.max()+1)
ax[0].axhline(0, color='grey', linewidth=0.5, linestyle='--')
ax[0].axvline(0, color='grey', linewidth=0.5, linestyle='--')

# Plot ReLU activated
ax[1].plot(x, relu(x))
ax[1].set_title('ReLU Activated')

ax[1].set_ylim(x.min()-1, x.max()+1)
ax[1].axhline(0, color='grey', linewidth=0.5, linestyle='--')
ax[1].axvline(0, color='grey', linewidth=0.5, linestyle='--')

plt.tight_layout()
plt.show()
```

### II. Dying ReLU Problem

While ReLU is an effective and commonly used activation function, there is a drawback to this function, called the "dying ReLU" problem. With ReLU, neurons can become inactive if they output a value of 0 for all inputs, effectively "dying" and preventing the neuron from learning any further. This can occur if the weighted sum of inputs to a neuron is consistently negative. This can lead to a portion of the network not learning anything, reducing the capacity and effectiveness of the model. We will introduce a solution to this problem with Leaky ReLU.

## B. Leaky ReLU

*Leaky ReLU* is designed to solve the dying ReLU problem by introducing an $\alpha$ parameter that will help make negative value into a much smaller negative number rather than making it completely zero. This parameter represents the slope function for negative input values. It is a small positive constant typically set to a very small value, such as 0.01. Below is the function:

$$
\text{LeakyReLU}(x) = \begin{cases}
x, & \text{if } x \geq 0 \\
\alpha x, & \text{otherwise}
\end{cases}
$$

### I. Visualizing Leaky ReLU

```python
import numpy as np
import matplotlib.pyplot as plt


def leaky_relu(x, alpha=0.1):
    """
    Leaky ReLU equation.
    """
    return np.where(x >= 0, x, alpha * x)


# Make linear array
x = np.linspace(-10, 10, 100)

# Init subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Plot Unactivated
ax[0].plot(x, x)
ax[0].set_title('Unactivated')

ax[0].set_ylim(x.min()-1, x.max()+1)
ax[0].axhline(0, color='grey', linewidth=0.5, linestyle='--')
ax[0].axvline(0, color='grey', linewidth=0.5, linestyle='--')

# Plot ReLU activated
ax[1].plot(x, leaky_relu(x))
ax[1].set_title('Leaky ReLU Activated')

ax[1].set_ylim(x.min()-1, x.max()+1)
ax[1].axhline(0, color='grey', linewidth=0.5, linestyle='--')
ax[1].axvline(0, color='grey', linewidth=0.5, linestyle='--')

plt.tight_layout()
plt.show()
```

## C. Sigmoid

*Sigmoid* maps the output of a neuron to a range between 0 and 1. This is used for hidden layers, as well as the output layer for a binary classification problem with an output size of 1. Below is the equations for sigmoid and its derivative:

$$ σ(x) = \frac{1}{1 + e^{-x}}\ $$

Below I will create the NumPy implementation of sigmoid and its derivative.

```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    """
    Sigmoid activation function.
    """
    return 1 / (1 + np.exp(-x))


# Generate an array of values ranging from -10 to 10
x = np.linspace(-10, 10, 100)

# Plot the original and sigmoid-transformed values
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, x)
plt.title("Original Values")

plt.subplot(1, 2, 2)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid Transformed Values")

plt.tight_layout()
plt.show()
```

## D. Hyperbolic Tangent (Tanh)

*Tanh* maps its input \(x\) to a value between -1 and 1. It is an S-shaped curve, similar to the sigmoid function, but with output values ranging from -1 to 1. The tanh function is symmetric around the origin and has zero mean, which can be beneficial in the context of centering and scaling data in neural networks. Below is the formula:

$$
\tanh(x) = \frac{{e^x - e^{-x}}}{{e^x + e^{-x}}}
$$

### I. Visualizing Tanh

We will use NumPy's `np.tanh()` function to visualize Tanh.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate an array of values ranging from -10 to 10
x = np.linspace(-10, 10, 100)

# Plot the original and tanh-transformed values
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, x)
plt.title("Original Values")

plt.subplot(1, 2, 2)
plt.plot(x, np.tanh(x))
plt.title("Tanh Transformed Values")

plt.tight_layout()
plt.show()
```

## E. Softmax

*Softmax* is used to transform a vector of real numbers into a probability distribution over multiple classes. This is typically applied to the last layer of a multiclass classification problem. Below is the formula:

$$
\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{N} e^{x_j}}
$$

### I. Visualizing Softmax

```python
import matplotlib.pyplot as plt
import numpy as np


def softmax(x):
    """
    Softmax Activation Function
    """
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)


output_logits = np.random.randn(5, 3)
softmax_logits = softmax(output_logits)

fig, ax = plt.subplots(3, 1, figsize=(8, 12))

# Plot raw logits
cax0 = ax[0].imshow(output_logits, cmap='coolwarm', aspect='auto')
ax[0].set_title('Raw Logits')
fig.colorbar(cax0, ax=ax[0])

# Plot softmax probabilities
cax1 = ax[1].imshow(softmax_logits, cmap='viridis', aspect='auto')
ax[1].set_title('Softmax Probabilities')
fig.colorbar(cax1, ax=ax[1])

# Plot sum of probabilities along the last axis
ax[2].plot(softmax_logits.sum(axis=-1), 'o-')
ax[2].set_title('Sum of Probabilities along Last Axis')
ax[2].set_xlabel('Sample Index')
ax[2].set_ylabel('Sum of Probabilities')

plt.tight_layout()
plt.show()
```

# 6 - Challenges in Neural Network Training

In this section, we will discuss the common challenges in training a neural network, and potential solutions to these problems.

## A. Overfitting vs Underfitting

Two of the biggest challenges in deep learning training is the problems of a model underfitting or overfitting to the training data.

### I. Overfitting

*Overfitting* is when the model learns to memorize the training data rather than that actual patterns in the data. This can result in very good results in training, but poor results to unseen data. Lets talk about some solutions to overfitting.

#### a. L1/L2 Regularization

1. **L1 Regularization (Lasso Regression)** - Adds the sum of the absolute values of all the weights in the model to the loss function. This shrinks the less important featue's coefficients to zero, which helps penalize large weights more severely, leading to sparse solutions where only a subset of weights are non-zero. This also introduces feature selection. Below is a mathmatical representaton:
$$
J_{\text{total}} = J(\hat{y}, y) + λ∑^n_{i=1}|w_i|
$$
where:
- $ J $ is the total loss function.
- $ J(\hat{y}, y) $ is the original loss function (such as mean squared error).
- $ \lambda $ is the regularization parameter.
- $ \sum_{i=1}^{n} |w_i| $ represents the sum of absolute values of all weights in the neural network.


2. **L2 Regularization (Ridge Regression)** - Adds the sum of squares of all weights in the model to the loss function. It allows for learning of complex patterns in the data. Unlike L1 regression, L2 does not introduce sparsity in the weights. It instead shrinks the weights toward zero without them actually being zero. Below is a mathmatical representaton:
$$
J_{\text{total}} = J(\hat{y}, y) + λ∑^n_{i=1}w^2_i
$$
where:
- $ J $ is the total loss function.
- $ J(\hat{y}, y) $ is the original loss function (such as mean squared error).
- $ \lambda $ is the regularization parameter.
- $ \sum_{i=1}^{n} w_i^2 $ represents the sum of squares of all weights in the neural network.



#### b. Dropout Layer

The *dropout layer* is a layer that during training, randomly zeros out a fraction $p$ of the proceding layer. Below is an image from [Wenkang Wei](https://wenkangwei.github.io/2020/11/13/DL-DropOut/) that represents using dropout during training:

<img src='https://wenkangwei.github.io/images/DL/dropout.jpg' width='800' height='300'>

It is important to note that dropout is only used during training, not evaluation. To compensate, we scale the outputs by a factor of $\frac{1}{1-p}$ during training.

#### c. Early Stopping

*Early stopping* is a technique used to prevent overfitting by stopping training early when the loss during validation begins to get worse. Below we can visualize this with a graph made by [Ananda Mohon Ghosh](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*nhmPdWSGh3ziatQKOmVq0Q.png):

<img src='https://miro.medium.com/v2/resize:fit:973/1*nhmPdWSGh3ziatQKOmVq0Q.png' width='500' height='300'>

Our validation set evaluates how effective our model is on unseen data. If the validation loss begins to decrease, it means our model is now begging to become less effective, suggesting a chance of overfitting.

#### d. Use More Data

To prevent overfitting, its important to collect as much data as we can for training. The more data there is, the more diverse patterns the model can learn from, decreasing the chance of overfitting to the training data.

### II. Underfitting

*Underfitting* is when the model is too simple to capture the underlying structure of the data, leading to poor training convergence.

#### a. Add Network Complexity

To deal with underfitting, we can make our model deeper and more complex by adding more layers and neurons, helping our model capture more complex patterns in the data.

## B. Vanishing and Exploding Gradients

Another challenge in deep neural network training is *vanishing* and *exploding* gradients. During backpropagation, gradients can become extremely small (vanishing) or large (exploding), making training difficult. Lets look at some solutions to this problem.

### I. Gradient Clipping

*Gradient clipping* is a technique in deep learning that helps prevent our gradients from exploding. The basic idea of gradient clipping is to set a threshold value, and if the norm of the gradients exceeds this threshold, the gradients are rescaled such that their norm is capped at the threshold. This helps to keep the gradients within a reasonable range, preventing them from growing too large.

### II. Batch Normalization

*Batch normalization* helps prevent both vanishing and exploding gradients by normalizing the input to a layer for each mini-batch during training. The normalization process involves subtracting the mini-batch mean and dividing by the mini-batch standard deviation. Given a mini-batch of inputs $ \mathbf{x} $ for a particular layer, batch normalization computes:

1. **Compute Mean and Variance**: Calculate the mean $ \mu_B $ and variance $ \sigma_B^2 $ of the mini-batch $ \mathbf{x} $:
$$
\mu_B = \frac{1}{m} \sum_{i=1}^{m} x_i
$$
$$
\sigma_B^2 = \frac{1}{m} \sum_{i=1}^{m} (x_i - \mu_B)^2
$$
Where $ m $ is the batch size, and $ x_i $ represents the $ i $-th input in the mini-batch.

2. **Normalize**: Normalize each input in the mini-batch using the mean and variance:
$$
\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}
$$
Where $ \epsilon $ is a small constant (typically $ 10^{-5} $) added for numerical stability to avoid division by zero.
3. **Scale and Shift**: Introduce learnable parameters $ \gamma $ (scale) and $ \beta $ (shift) to linearly transform the normalized activations:
$$
y_i = \gamma \hat{x}_i + \beta
$$
Where $ \gamma $ and $ \beta $ are learnable parameters specific to each layer.


### III. Weight Initialization

Effective weight initialization may help prevent issues with exploding and vanishing gradients. Below are the popular weight initialization techniques in addition to random normal and uniform initialization:

1. **Xavier Initialization (Glorot Initialization)** - This method initializes the weights of the network using a Gaussian distribution with zero mean and a variance calculated based on the number of input and output units of the layer. It aims to keep the variance of activations and gradients roughly constant across layers. This is designed for sigmoid and tanh activation functions.

2. **He Initialization** - This technique, proposed by Kaiming He et al., is specifically designed for ReLU activation functions. It initializes the weights using a Gaussian distribution with zero mean and a variance scaled based on the number of input units, aiming to maintain a constant variance of activations throughout the network.

3. **Orthogonal Initialization** - This approach initializes the weights as orthogonal matrices, ensuring that the gradients neither vanish nor explode as they propagate through the network.


### IV. Use ReLU and Leaky ReLU

Some activation functions, like tanh and sigmoid have derivatives that approach zero as their inputs move towards extreme values (very large positive or very large negative). This means that if the input to these functions becomes too large or too small, the derivative becomes extremely close to zero. When the derivative is close to zero, it implies that the gradient, which is crucial for updating the weights during training using backpropagation, becomes very small, leading to the vanishing gradient problem. To deal with the issues of vanishing gradients, researchers have introduced ReLU and Leaky ReLU. These activation functions have larger or constant ranges where their derivatives are nonzero.



## C. Computational Resources

Training deep neural newtorks can be computationaly and time expensive, especially with large datasets or complex netorks. To deal with these issues, we can utilize GPUs and TPUs during training, which improves training time. You can also distribute training across multiple devices or machines to help speed up training.

## D. Learning Rate Selection

Choosing the correct learning rate is a challenge when it comes to neural network training. A learning rate too high may overshoot the minimum of the loss function and fail to converge, and a learning rate too low will cause the training process to be very slow, or the model may get stuck in a local minima and struggle to escape it due to the small step size. A common thought is to start with a relativly lower learning rate to make sure we can have a steady optimization for each step without overshooting. This can be a good technique, but what if we want to optimize this even further to deal with long training time issues? We can optimize this with a learning rate scheduler.

#### I. Learning Rate Scheduler

A learning rate schedule can help optimize our training process by dynamically changing our learning rate overtime. Some common learning rate schedules are:

1. **Step Decay**: This schedule reduces the learning rate by a factor at predefined epochs. Below is the formula:
$$
\text{lr}(epoch) = \text{initial_lr} \times \text{factor}^{\lfloor \frac{\text{epoch}}{\text{step_size}} \rfloor}
$$
Where:
- `initial_lr`: Initial learning rate.
- `factor`: Factor by which the learning rate will be reduced.
- `step_size`: Number of epochs after which the learning rate will be reduced.

2. **Exponential Decay:** This schedule reduces the learning rate exponentially over time. Below is the formula:
$$
\text{lr}(epoch) = \text{initial_lr} \times \text{exp}(-\text{decay_rate} \times \text{epoch})
$$
Where:
- `initial_lr`: Initial learning rate.
- `decay_rate`: Rate of decay for the learning rate.

3. **Cosine Annealing**: This schedule varies the learning rate according to the cosine function over the course of training epochs. Below is the formula:
$$
\text{lr}(epoch) = \text{initial_lr} \times \frac{1}{2} \left(1 + \text{cos}\left(\frac{\text{epoch}}{\text{max_epochs}} \times \pi \right) \right)
$$
Where:
- `initial_lr`: Initial learning rate.
- `max_epochs`: Total number of training epochs.

4. **ReduceLROnPlateau**: This schedule reduces the learning rate when a monitored metric has stopped improving.
